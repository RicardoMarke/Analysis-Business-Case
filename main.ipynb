## Segmento com maior percentual

### Agrupando por segmento e ano, somando vendas na variável "vendas_por_segmento"

```python
vendas_por_segmento = df.groupby(['Segment', 'Ano'])['Sales ($)'].sum().reset_index().sort_values('Ano')
print(vendas_por_segmento)
```

### Pivotando o dataframe

```python
vendas_pivot = vendas_por_segmento.pivot(index='Segment', columns='Ano', values='Sales ($)')
```

### Exibindo a tabela pivot criada

```python
vendas_pivot
```

## Calculando o crescimento percentual entre 2014 a 2017

```python
vendas_pivot['Crescimento Percentual'] = ((vendas_pivot[2017] - vendas_pivot[2014]) / vendas_pivot[2014]) * 100
```

### Identificando o segmento com o maior crescimento percentual

```python
segmento_maior_crescimento = vendas_pivot['Crescimento Percentual'].idxmax()
maior_crescimento_valor = vendas_pivot['Crescimento Percentual'].max()
```

```python
print(segmento_maior_crescimento)
print(maior_crescimento_valor)
```

### Plotando o gráfico

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=vendas_pivot.reset_index(), x='Segment', y='Crescimento Percentual', palette='viridis')
plt.title('Crescimento Percentual por Segmento de 2014 a 2017')
plt.xlabel('Segmento')
plt.ylabel('Crescimento Percentual (%)')
plt.ylim(0, vendas_pivot['Crescimento Percentual'].max() + 10)  # Define o limite do eixo y
plt.axhline(0, color='gray', linewidth=0.8, linestyle='--')  # Linha de referência no zero
plt.show()
```

## Projetando cresciemento de vendas deste segmento para os próximos 5 vanos

#### Filtrando apenas os anos de interesse

```python
anos_interesse = [2014, 2015, 2016, 2017]
vendas_filtradas = vendas_por_segmento[vendas_por_segmento['Ano'].isin(anos_interesse)]
vendas_filtradas
```

#### Agrupando por Segment e Ano e somando as vendas

```python
vendas_por_segmento = df.groupby(['Segment', 'Ano'])['Sales ($)'].sum().reset_index().sort_values('Ano')
print(vendas_por_segmento)
```

#### Pivotando o DataFrame para facilitar o cálculo

```python
vendas_pivot = vendas_filtradas.pivot(index='Segment', columns='Ano', values='Sales ($)')
```

```python
vendas_pivot
```

#### Calculando o crescimento percentual de 2016 para 2017 e convertendo para taxa decimal

```python
crescimento_percentual = ((vendas_pivot.loc['Corporate', 2017] - vendas_pivot.loc['Corporate', 2016]) / vendas_pivot.loc['Corporate', 2016]) * 100
taxa_crescimento = crescimento_percentual / 100
```

#### Projeção para os próximos 5 anos

```python
vendas_projetadas = [vendas_pivot.loc['Corporate', 2017]]  # Começando com o valor de 2017
for i in range(1, 6):  # Para os próximos 5 anos
    proxima_venda = vendas_projetadas[-1] * (1 + taxa_crescimento)
    vendas_projetadas.append(proxima_venda)
```

#### Anos de projeção

##### De 2017 a 2022

```python
anos_projecao = [2017 + i for i in range(6)]
```

#### Criando um DataFrame para as vendas projetadas

```python
df_projecao = pd.DataFrame({'Ano': anos_projecao, 'Vendas Projetadas': vendas_projetadas})
```

```python
df_projecao
```

#### Plotando o gráfico

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df_projecao, x='Ano', y='Vendas Projetadas', marker='o')
plt.title('Projeção de Vendas para o Segmento Corporate nos Próximos 5 Anos')
plt.xlabel('Ano')
plt.ylabel('Vendas Projetadas ($)')
plt.xticks(anos_projecao)  # Define os ticks do eixo x
plt.grid(True)
plt.show()
```
